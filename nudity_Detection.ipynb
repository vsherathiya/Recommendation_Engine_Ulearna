```python
import os
import csv
# Import your prediction module

def classify_images_in_folder(folder_path, output_csv):
    # List all image files in the folder
    image_files = [f for f in os.listdir(folder_path) if f.endswith(('.jpg', '.jpeg', '.png'))]

    # Open a CSV file to write the results
    with open(output_csv, 'w', newline='') as csvfile:
        fieldnames = ['image', 'drawings', 'hentai', 'neutral', 'porn', 'sexy']
        writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
        writer.writeheader()

        # Process each image file
        for image_file in image_files:
            image_path = os.path.join(folder_path, image_file)
            try:
                # Predict and get the scores
                result = predict.classify(model, image_path)
                # Write the results to the CSV file
                writer.writerow({
                    'image': image_file,
                    **result[image_path]  # Unpack the scores into the row
                })
            except Exception as e:
                print(f"Error processing {image_file}: {e}")

# Usage
folder_path = 'D:\\Recommendation_Engine\\Frames\\hentai2'  # Path to your folder
output_csv = 'output_predictionshentai2.csv'  # Path to the output CSV file
classify_images_in_folder(folder_path, output_csv)
```

```python
import cv2
import os
# from your_prediction_module import predict  # Import your prediction module

def extract_frames(video_path, interval=1):
    """
    Extracts frames from a video file at specified intervals.
    :param video_path: Path to the video file.
    :param interval: Interval in seconds between frames to extract.
    :return: List of frame images.
    """
    cap = cv2.VideoCapture(video_path)
    if not cap.isOpened():
        print(f"Failed to open video: {video_path}")
        return []

    fps = cap.get(cv2.CAP_PROP_FPS)
    print(f"FPS: {fps}")
    frames = []

    while cap.isOpened():
        ret, frame = cap.read()
        if not ret:
            print("Reached end of video or failed to read frame.")
            break

        current_frame = int(cap.get(cv2.CAP_PROP_POS_FRAMES))
        if current_frame % (int(fps) * interval) == 0:
            frames.append(frame)

    cap.release()
    return frames

def classify_video_frames(video_path, interval=1):
    frames = extract_frames(video_path, interval)
    print(frames)
    nudity_detected = True

    for frame in frames:
        # Apply your prediction model to the frame
        result = predict.classify(model, frame)
        print(result)
        # Check if nudity is detected in the frame
        if result['neutral'] > 0.95:  # Adjust thresholds as needed
            nudity_detected = False
            break

    return nudity_detected

# Usage
video_path = r'D:\Recommendation_Engine\Sexy\THIS Is VERY Useful! .mp4'  # Path to your video file
nudity_present = classify_video_frames(video_path)
print(f"Nudity Present: {nudity_present}")
```

```python
import cv2
import numpy as np
import tensorflow as tf
import tensorflow_hub
from nsfw_detector import predict
import tempfile
import os
import time

def load_model(model_path):
    """
    Load the pre-trained TensorFlow model.
    """
    return tf.keras.models.load_model(
        model_path, 
        custom_objects={"KerasLayer": tensorflow_hub.KerasLayer}
    )


def classify_frame(frame, model):
    # Preprocess the frame (resize, normalize, etc.)
    resized_frame = cv2.resize(frame, (224, 224))  # Example size
    normalized_frame = resized_frame / 255.0

    # Create a temporary file and close it to ensure it's not locked
    temp_file, temp_filename = tempfile.mkstemp(suffix='.jpg')
    os.close(temp_file)

    # Write frame to temporary file
    cv2.imwrite(temp_filename, normalized_frame * 255)  # Convert back to range [0, 255] if necessary

    # Classify the frame using the model
    result = predict.classify(model, temp_filename)

    # Delete the temporary file
    os.remove(temp_filename)

    return result


def process_video(video_path, model):
    """
    Process a video, classifying each frame for nudity and aggregate results.
    """
    cap = cv2.VideoCapture(video_path)
    flagged_frames = 0
    total_frames = 0

    while cap.isOpened():
        ret, frame = cap.read()
        if not ret:
            break

        total_frames += 1

        # Classify the frame
        result = classify_frame(frame, model)

        # Interpret the result (adjust according to your model's output)
        if result.get('nudity', 0) > 0.5:  # Threshold, adjust as needed
            flagged_frames += 1

    cap.release()
    cv2.destroyAllWindows()

    # Aggregate results
    nudity_presence = flagged_frames > 0
    nudity_percentage = (flagged_frames / total_frames) * 100 if total_frames > 0 else 0

    return {
        "nudity_detected": nudity_presence,
        "nudity_percentage": nudity_percentage,
        "total_frames": total_frames,
        "flagged_frames": flagged_frames
    }

# def process_video(video_path, model):
#     """
#     Process a video, classifying each frame for nudity and aggregate results.
#     """
#     cap = cv2.VideoCapture(video_path)
#     consecutive_non_neutral_frames = 0
#     max_consecutive_non_neutral = 0  # Track the maximum number of consecutive non-neutral frames
#     nudity_detected = False

#     while cap.isOpened():
#         ret, frame = cap.read()
#         if not ret:
#             break

#         # Classify the frame
#         result = classify_frame(frame, model)

#         # Check if the frame is non-neutral (adjust according to your model's output)
#         if result.get('neutral', 1) < 0.5:  # Assuming 'neutral' probability less than 0.5 means non-neutral
#             consecutive_non_neutral_frames += 1
#             max_consecutive_non_neutral = max(consecutive_non_neutral_frames, max_consecutive_non_neutral)
#         else:
#             consecutive_non_neutral_frames = 0

#         # Check if we have 3 to 5 consecutive non-neutral frames
#         if max_consecutive_non_neutral >= 3:
#             nudity_detected = True
#             break  # Break the loop if nudity is detected

#     cap.release()
#     cv2.destroyAllWindows()

#     return {
#         "nudity_detected": nudity_detected,
#         "max_consecutive_non_neutral": max_consecutive_non_neutral
#     }
```

```python
import cv2
import os

def save_video_frames(video_path, save_folder, max_fps):
    # Create a folder to save the frames
    if not os.path.exists(save_folder):
        os.makedirs(save_folder)

    # Open the video
    cap = cv2.VideoCapture(video_path)
    if not cap.isOpened():
        print("Error: Could not open video.")
        return

    # Get frame rate of the video
    actual_fps = cap.get(cv2.CAP_PROP_FPS)

    # Calculate the frame skipping rate
    skip_rate = max(1, int(round(actual_fps / max_fps)))

    frame_count = 0
    saved_frame_count = 0

    # Read through the video
    while True:
        ret, frame = cap.read()
        if not ret:
            break

        # Skip frames to meet the max_fps requirement
        if frame_count % skip_rate == 0:
            # Save the frame
            frame_path = os.path.join(save_folder, f"frame_{saved_frame_count}.jpg")
            cv2.imwrite(frame_path, frame)
            saved_frame_count += 1

        frame_count += 1

    cap.release()

# Example usage
video_path = r"D:\Recommendation_Engine\Sexy\Spaghetti anime part 5 .mp4"  # Update this path
save_folder = r"D:\Recommendation_Engine\Frames\hentai2"
os.makedirs(save_folder,exist_ok=True)# Update this path
max_fps = 10  # Maximum FPS to be considered
save_video_frames(video_path, save_folder, max_fps)
```
